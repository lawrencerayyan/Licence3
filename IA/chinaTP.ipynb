```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
```

```python
#On importe l'image

from sklearn.datasets import load_sample_image
image = load_sample_image("china.jpg")


#Si vous ne voulez pas d'affichage sur l'axe x et l'axe y
plt.axes(xticks=[], yticks=[])

#pour afficher les images
plt.imshow(image)
```

```python
#Dimension de l'image original
image.shape
```

Combien de pixels y a-t-il ?

Quelle est la dimension de l'espace de couleurs?

```python


rows  = image.shape[0]

cols= image.shape[1]

#on aplatit  l'image
image_p = image.reshape(rows*cols,3)

image_p.shape

```

```python
print(image_p[0,:])

#Que reprsente l'affichage?
```

Comparez les rsultats obtenus par image.shape et image_p.shape

Que reprsente image_p?

```python
#Normalisation des couleurs 

image_p=image_p/255

print(image_p[0,:])

#Que reprsente l'affichage?
```

Afficher 50 premiers lments de image_p.
Que constatez-vous?

Les trois dimensions sont des intensits des couleurs RGB (Red Green Blue)

L'intensit de chaque couleur est cod par 0-255, il y aura donc 2^8 intensit  par couleur.  

Par pixel, il y aura (2^8)^3= 2^24=16777216 couleurs diffrentes


Nous ferons clustering pour regrouper les couleurs similaires par la mthode de KMeans

(Nous comprssons l'image) 


```python
from sklearn.cluster import KMeans

model2 =KMeans(2)

model2.fit(image_p)


centers= model2.cluster_centers_
labels = model2.labels_

```

Que reprsentent  les variables centers et labels? 

Affichez centers

Affichez 100 premieres valeurs de labels

Affichez 100 dernieres valeurs de centers

Quelles sont les couleurs utilises.

```python
# Nous reconstruisons  l'image avec 2 couleurs

new_colors =[ centers[l] for l in labels]

img = np.reshape(new_colors, image.shape)
```

Que reprsentent 

- centers[labels]

- new_colors

- Pourquoi on fait np.reshape(new_colors, image.shape)

```python
#dessinez l'image  avec 2 couleurs

print(img.shape)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize= (16,6))
#ax1.axis('off')
ax1.set_title('Original image ')
ax1.imshow(image)
#ax2.axis('off')
ax2.set_title('Quantized image (K=2 colors)')
ax2.imshow(img)
plt.show()
```
