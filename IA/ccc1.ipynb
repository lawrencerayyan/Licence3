```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Question 1
wine_data = load_wine()

# Question 2
labels = wine_data.target_names

# Question 3
first_attribute_description = wine_data.feature_names[0]

# Question 4
label_15th_data = wine_data.target[14]
 
# Question 5
X = wine_data.data
y = wine_data.target

# Question 6
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)

# Question 7
knn_classifier = KNeighborsClassifier(n_neighbors=2)
knn_classifier.fit(X_train, y_train)

# Question 8
nouveau = [[16, 3.1, 1.3, 24, 218, 2.5, 2.4, 0.27, 1.50, 3.4, 1.3, 3, 480]]
prediction_knn = knn_classifier.predict(nouveau)

# Question 9
accuracy_knn = accuracy_score(y_test, knn_classifier.predict(X_test))

# Question 10
labels_first_5_test_data = y_test[:5]
predicted_labels_first_5_test_data = knn_classifier.predict(X_test[:5])

# Question 11
decision_tree_classifier = DecisionTreeClassifier(max_depth=3)
decision_tree_classifier.fit(X_train, y_train)

# Question 12
prediction_tree = decision_tree_classifier.predict(nouveau)

# Question 13
accuracy_tree = accuracy_score(y_test, decision_tree_classifier.predict(X_test))

# Question 14
best_method = "KNeighborsClassifier" if accuracy_knn > accuracy_tree else "DecisionTreeClassifier"


print("Labels possibles pour une donnée:", labels)
print("Représentation du premier attribut:", first_attribute_description)
print("Étiquette de la 15ème donnée:", label_15th_data)
print("\nMatrice de données X:\n", X)
print("\nVecteur d'étiquettes y:\n", y)
print("\nEnsemble d'apprentissage X_train:\n", X_train)
print("\nEnsemble de test X_test:\n", X_test)
print("\nScore de précision (accuracy) pour KNeighborsClassifier:", accuracy_knn)
print("\nLabels des 5 premières données de test:", labels_first_5_test_data)
print("Labels prédits pour les 5 premières données de test (KNeighborsClassifier):", predicted_labels_first_5_test_data)
print("\nScore de précision (accuracy) pour DecisionTreeClassifier:", accuracy_tree)
print("Meilleure méthode:", best_method)
```
